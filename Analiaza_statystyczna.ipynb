```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Załaduj dane (np. Titanic)
df = pd.read_csv('titanic.csv')

# Wyciągnięcie statystyk dla zmiennej Fare
fare_stats = df['Fare'].describe()
print(fare_stats)

# Wyciągnięcie konkretnych wartości:
mean_fare = df['Fare'].mean()
median_fare = df['Fare'].median()
q1 = df['Fare'].quantile(0.25)
q3 = df['Fare'].quantile(0.75)
std_fare = df['Fare'].std()

print(f"Średnia: {mean_fare:.2f}")
print(f"Mediana: {median_fare:.2f}")
print(f"1. kwartyl (Q1): {q1:.2f}")
print(f"3. kwartyl (Q3): {q3:.2f}")
print(f"Odchylenie standardowe: {std_fare:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Fare'])
plt.title('Wykres pudełkowy opłaty (Fare)')
plt.xlabel('Fare')
plt.show()
```
